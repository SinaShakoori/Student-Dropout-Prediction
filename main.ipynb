dataset: https://www.kaggle.com/datasets/thedevastator/higher-education-predictors-of-student-retention

```python
import pandas as pd
```

```python
dataset = pd.read_csv("dataset.csv")
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset['Target'].value_counts()
```

```python
dataset = dataset[dataset['Target'] != 'Enrolled']
```

```python
dataset['Target'].value_counts()
```

```python

Y =(dataset.iloc[:,34:])
```

```python
X = dataset.drop(dataset.columns[34],axis=1)
```

```python
X.shape
```

```python
X.head()
```

```python
Y.shape
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
Y = le.fit_transform(Y["Target"])
```

```python
le.classes_
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
model = LogisticRegression(max_iter=800)
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)

print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(class_weight='balanced',n_estimators=150)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

XGBOOST

```python
from xgboost import XGBClassifier
```

```python
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```
